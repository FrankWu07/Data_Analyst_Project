# Guided Project_ Star Wars Survey

While waiting for [Star Wars: The Force Awakens](https://en.wikipedia.org/wiki/Star_Wars:_The_Force_Awakens) to come out, the team at [FiveThirtyEight](https://fivethirtyeight.com/) became interested in answering some questions about Star Wars fans. In particular, they wondered: **does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?**

The team needed to collect data addressing this question. To do this, they surveyed *Star Wars* fans using the online tool SurveyMonkey. They received 835 total responses, which you download from [their GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

## 1.Read Data

```python
## read and encoding data
import pandas as pd
star_wars=pd.read_csv('star_wars.csv', encoding='"ISO-8859-1"')
star_wars.head(10)
```

```python
star_wars.columns
```

The data has several columns, including:

* `RespondentID` - An anonymized ID for the respondent (person taking the survey)
* `Gender` - The respondent's gender
* `Age` - The respondent's age
* `Household Income` - The respondent's income
* `Education` - The respondent's education level
* `Location (Census Region)` - The respondent's location
* `Have you seen any of the 6 films in the Star Wars franchise?` - Has a `Yes` or `No` response
* `Do you consider yourself to be a fan of the Star Wars film franchise?` - Has a `Yes` or `No` response

There are several other columns containing answers to questions about the Star Wars movies. For some questions, the respondent had to check one or more boxes. This type of data is difficult to represent in columnar format. As a result, this data set needs a lot of cleaning.

## 2. Clean  And Analysis Data


```python
## remove NaN from 'RespondentID' column and keep unique 'respondentID'
num_null_C1=star_wars['RespondentID'].isnull().sum()
print('The number of None value in RespondentID is: {0}'.format(num_null_C1))
star_wars=star_wars[pd.notnull(star_wars['RespondentID'])]
print('\n')
num_null_C1=star_wars['RespondentID'].isnull().sum()
print('The number of None value in RespondentID is: {0}'.format(num_null_C1))
print('\n')
star_wars.head(10)
```

```python
## Convert two columns to the Boolean type.
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False))
print('\n')
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False))

convert_dic={'Yes': True, 
             'No': False
            }
convert_cols=['Have you seen any of the 6 films in the Star Wars franchise?',
              'Do you consider yourself to be a fan of the Star Wars film franchise?']
for c in convert_cols:
    star_wars[c]= star_wars[c].map(convert_dic)
```

```python
star_wars.head()
```

We have been convert 'Yes' and 'No' to 'True' and 'False' in above two columns.

The next six columns represent a single checkbox question. The respondent checked off a series of boxes in response to the question, `Which of the following Star Wars films have you seen? Please select all that apply`.

The columns for this question are:

* `Which of the following Star Wars films have you seen? Please select all that apply`. - Whether or not the respondent saw `Star Wars: Episode I The Phantom Menace`.
* `Unnamed: 4` - Whether or not the respondent saw `Star Wars: Episode II Attack of the Clones`.
* `Unnamed: 5` - Whether or not the respondent saw `Star Wars: Episode III Revenge of the Sith`.
* `Unnamed: 6` - Whether or not the respondent saw `Star Wars: Episode IV A New Hope`.
* `Unnamed: 7` - Whether or not the respondent saw `Star Wars: Episode V The Empire Strikes Back`.
* `Unnamed: 8` - Whether or not the respondent saw `Star Wars: Episode VI Return of the Jedi`.

For each of these columns, if the value in a cell is the name of the movie, that means the respondent saw the movie. If the value is `NaN`, the respondent either didn't answer or didn't see the movie. We'll assume that they didn't see the movie.

We'll need to convert each of these columns to a Boolean, convert `move name` to `True`, and `NaN` to `False`. 

```python
## convert each of these columns to a Boolean, convert move name to True, and NaN to False
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

star_wars.head()
```

```python
## rename above columns
rename_col={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
}
star_wars = star_wars.rename(columns=rename_col)
pd.set_option('display.max_columns', None)
star_wars.head()
```

The next six columns ask the respondent to rank the Star Wars movies in order of least favorite to most favorite. 1 means the film was the most favorite, and 6 means it was the least favorite. Each of the following columns can contain the value 1, 2, 3, 4, 5, 6, or NaN:
* `Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film`. - How much the respondent liked `Star Wars: Episode I The Phantom Menace`
* `Unnamed: 10` - How much the respondent liked `Star Wars: Episode II Attack of the Clones`
* `Unnamed: 11` - How much the respondent liked `Star Wars: Episode III Revenge of the Sith`
* `Unnamed: 12` - How much the respondent liked `Star Wars: Episode IV A New Hope`
* `Unnamed: 13` - How much the respondent liked `Star Wars: Episode V The Empire Strikes Back`
* `Unnamed: 14` - How much the respondent liked `Star Wars: Episode VI Return of the Jedi`

Fortunately, these columns don't require a lot of cleanup. We'll need to convert each column to a numeric type, though, then rename the columns so that we can tell what they represent more easily.

```python
## convert to float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars[star_wars.columns[9:15]].info()
```

```python
## rename the above columns
rename_cols2={
     "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
}
star_wars=star_wars.rename(columns=rename_cols2)
star_wars.head()
```

Now that we've cleaned up the ranking columns, we can find the highest-ranked movie more quickly. And generate a bar chart to display the result.

```python
## cacluate the mean value to each ranking
mean_ranking=star_wars[star_wars.columns[9:15]].mean()
mean_ranking
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

ax=mean_ranking.plot.bar(figsize=(8,6))
ax.set_xlabel('Movies')
ax.set_ylabel('Average Rankings')
ax.set_title('Six Star War Movies Average Rankings')
plt.show()
```

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As I suspected, it looks like the newer movies are rated much more highly than the older ones (lower ranking is better!). And `Star Wars: Episode V The Empire Strikes Back` is the most popular movie!

Earlier in this project, we cleaned up the seen columns and converted their values to the Boolean type. we can figure out how many people have seen each movie just by taking the sum of the column (even though they contain Boolean values).

```python
seen_sum=star_wars[star_wars.columns[3:9]].sum()
seen_sum
```

```python
ax=seen_sum.plot.bar(figsize=(8,6))
ax.set_xlabel('Movies')
ax.set_ylabel('Total People Seen Movie')
ax.set_title('Total numbers ')
plt.show()
```

We can find the new movie have more respondents and Star Wars: Episode V The Empire Strikes Back is the most popular movie!

### Which character shot first?

In the original 1977 movie Star Wars there was a scene where a smuggler Han shoots a bounty hunter Greedo. In the later releases thought the scene was change so that Greedo shoots first and only then Han shoots him as an act of self-defense. This scene has created a debate on whether it was necessary to change the scene and which alternative is more canonical. Let's look what the American society thinks about this alternation. We will also look at how it's correlated with the respondent age: may be younger viewers think it was Greedo who shot first because they did not see the original movie?([Source](https://en.wikipedia.org/wiki/Han_shot_first#:~:text=%22Han%20shot%20first%22%20refers%20to,scene%2C%20Han%20shoots%20Greedo%20dead.))

```python
star_wars['Which character shot first?'].value_counts(normalize=True)
```

About 40% of the respondees share the view that Han shot first, whereas less than 25% do not understand the question. Since this question referrs to a specific episode in the series, let's see if the people who did not understand the question have seen that particular episode.

```python
who=pd.pivot_table(star_wars, index='Which character shot first?', values=star_wars.columns[3:9], aggfunc=np.sum)
who=who.transpose()
who
```

```python
ax=who.plot.barh(figsize=(10,8), fontsize=12)
ax.set_xlabel('Total People')
ax.set_ylabel('Movies')
ax.set_title('Viewers by movie, depending on who shot first')
```

Since the "Who shot first" debate refers to `Episode 4: A New Hope`, we would expect that the people who did not understand the questions, would have seen that one the least. However, this is not the case. That category of respondents has seen `Episode 3: Revenge of the Sith the least`.

### Characters Respondents 

Let's clean up the columns in which the respondees rank the characters of the movies and afterwards find out which characters and most and least liked.

```python
## Check the columns
star_wars.columns[15:29]
```

```python
star_wars['Unnamed: 16'].value_counts()
```

We can see that favorite level contains:
* `Very favorably`
* `Somewhat favorably`
* `Neither favorably nor unfavorably (neutral)`
* `Somewhat unfavorably`
* `Unfamiliar (N/A)`
* `Very unfavorably`

```python
## Let's rename the Unname columns
chars = {
    'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.': 'Han Solo',
    'Unnamed: 16':'Luke Skywalker',
    'Unnamed: 17': 'Princess Leia Organa',
    'Unnamed: 18': 'Anakin Skywalker',
    'Unnamed: 19': 'Obi Wan Kenobi',
    'Unnamed: 20': 'Emperor Palpatine',
    'Unnamed: 21': 'Darth Vader',
    'Unnamed: 22': 'Lando Calrissian',
    'Unnamed: 23': 'Boba Fett',
    'Unnamed: 24': 'C-3P0',
    'Unnamed: 25': 'R2-D2',
    'Unnamed: 26': 'Jar Jar Binks',
    'Unnamed: 27': 'Padme Amidala', 
    'Unnamed: 28': 'Yoda'
    }
star_wars = star_wars.rename(columns=chars)
star_wars.columns[15:29]
```

We will convert the ranking system to numbers so that we can make calculations:

* `Very favorably` - for each rating the character will receive 2 points
* `Somewhat favorably` - for each rating the character will receive 1 point
* `Neither favorably nor unfavorably (neutral)` - 0 points
* `Somewhat unfavorably` - for each rating the character will be deducted 1 point
* `Very unfavorably` - for each rating the character will be deducted 2 points
* `Unfamiliar (N/A)` - 0 points

```python
## Mapping points
points={
    'Very favorably': 2,
    'Somewhat favorably': 1,
    'Neither favorably nor unfavorably (neutral)': 0,
    'Somewhat unfavorably': -1,
    'Very unfavorably': -2,
    'Unfamiliar (N/A)': 0,
    np.NaN: 0
}

characters = ['Han Solo', 'Luke Skywalker', 'Princess Leia Organa',
       'Anakin Skywalker', 'Obi Wan Kenobi', 'Emperor Palpatine',
       'Darth Vader', 'Lando Calrissian', 'Boba Fett', 'C-3P0', 'R2-D2',
       'Jar Jar Binks', 'Padme Amidala', 'Yoda']

for char in characters:
    star_wars[char] = star_wars[char].map(points)

```

```python
fav_movie=star_wars.loc[:,characters]
fav_movie=fav_movie.sum()
fav_movie
```

```python
ax=fav_movie.plot.barh(figsize=(10,8))
ax.set_title('Overall favorite characters')
ax.axvline(c='k')
y=fav_movie
for i, v in enumerate(y):
    ax.text(v, i , str(v), fontweight='bold')
plt.show()
```

Observation:

1) we could see that the most favorably character are Han Solo, Luke S, Princess, Obi Kenobi, R2D2 and yoda. 

2) The most disliked Characters are Jar Jar Binks, Darth Vader and Emperor.

3) The most controversial(spliting the likes and dislikes goes to) Lando and Anakin S.
